## Registro del 2024

```python
df_2024 = pd.read_excel('C:\\Users\\DELL\\Downloads\\ParqueVehicular2024\\Parque Vehicular.xlsx', sheet_name='Stock vehiculos', skiprows=9)
```

```python
df_2024.head()
```

```python
df_2024 = df_2024.iloc[1:19]
```

```python
df_2024.reset_index(drop=True, inplace=True)
```

```python
df_2024.columns = [
    'Tipo_Vehiculo',
    "Diciembre_2023", 
    "Enero", 
    "Febrero", 
    "Marzo", 
    "Abril", 
    "Mayo", 
    "Junio", 
    "Julio", 
    "Agosto", 
    "Septiembre", 
    "Octubre", 
    'Noviembre'
]
```

```python
df_2024['Tipo_Vehiculo'] = df_2024['Tipo_Vehiculo'].str.strip()
```

```python
filtro_automovil = df_2024['Tipo_Vehiculo'].str.contains('Automóvil')
suma_automovil = df_2024.loc[filtro_automovil, 'Noviembre'].sum()
df_historico.loc[df_historico['Tipo'] == 'Automóviles', '2024'] = suma_automovil
```

```python
df_historico
```

```python
filtro_autobus = df_2024['Tipo_Vehiculo'].str.contains('Autobús')
suma_autobus = df_2024.loc[filtro_autobus, 'Noviembre'].sum()
df_historico.loc[df_historico['Tipo'] == 'Autobuses', '2024'] = suma_autobus
df_historico
```

```python
df_historico.loc[df_historico['Tipo'] == 'Jeep', '2024'] = df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Jeep', 'Noviembre'].values[0]
df_historico.loc[df_historico['Tipo'] == 'Carga', '2024'] = df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Carga', 'Noviembre'].values[0]
df_historico.loc[df_historico['Tipo'] == 'Motocicletas', '2024'] = df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Motocicleta', 'Noviembre'].values[0]
df_historico.loc[df_historico['Tipo'] == 'Volteo', '2024'] = df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Volteo', 'Noviembre'].values[0]
df_historico.loc[df_historico['Tipo'] == 'Máquinas Pesadas', '2024'] = df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Maquina Pesada', 'Noviembre'].values[0]
```

```python
df_historico.loc[df_historico['Tipo'] == 'Otros', '2024'] = (df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Ambulancia', 'Noviembre'].values[0] + 
                                                             df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Fúnebre', 'Noviembre'].values[0] + 
                                                             df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Montacarga', 'Noviembre'].values[0] + 
                                                             df_2024.loc[df_2024['Tipo_Vehiculo'] == 'Remolque', 'Noviembre'].values[0])
```

```python
df_historico.iloc[:]
```

```python
df_historico.loc[df_historico['Tipo'] == 'Total', '2024'] = df_2024['Noviembre'].iloc[-1]
```

```python
df_historico['2024'] = df_historico['2024'].astype(int)
```
